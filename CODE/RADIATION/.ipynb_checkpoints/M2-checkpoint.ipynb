## DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('RADIATION.csv')
del df["UNIXTime"]
del df["Data"]
del df["Time"]
del df["TimeSunRise"]
del df["TimeSunSet"]
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df['Pressure'],color='red')

plt.subplot(1,2,2)
plt.hist(df['Humidity'],color='green')
```

```python
df.hist(figsize=(15,55),layout=(15,4), color='blue')
plt.show()
```

```python
df['Radiation'].hist(figsize=(10,5),color='yellow')
```

```python
sns.ecdfplot(df['Pressure'], color='orange') # scatter, plot, triplot, stackplot
```

```python
sns.kdeplot(df['Pressure'], color='purple') 
```

```python
df['Speed'].plot(kind='density')
```

```python
sns.displot(df['Humidity'], color='purple') 
# barplot, boxenplot, boxplot, countplot, displot, distplot, ecdfplot, histplot, kdeplot, pointplot, violinplot, stripplot
```

```python
sns.ecdfplot(df['WindDirection(Degrees)'], color='coral') # residplot, scatterplot 
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),annot = True, fmt='0.2%',cmap = 'autumn',ax=ax)
```
