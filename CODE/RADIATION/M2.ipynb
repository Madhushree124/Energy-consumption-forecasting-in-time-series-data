```python
# Time series plot
plt.figure(figsize=(12, 6))
plt.plot(df['UNIXTime'], df['Radiation'], label='Solar Radiation')
plt.xlabel('UNIX Time')
plt.ylabel('Solar Radiation')
plt.title('Solar Radiation Over Time')
plt.legend()
plt.show()
```

```python
# Box plot by hour
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Hour', y='Radiation')
plt.xlabel('Hour')
plt.ylabel('Solar Radiation')
plt.title('Solar Radiation by Hour of the Day')
plt.show()
```
